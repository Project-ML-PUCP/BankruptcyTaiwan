### **2. Lectura Inicial de datos**

#### 2.1. Carga de datos

```python
# Se carga el dataset normalizado
df_bankruptcy = pd.read_pickle(os.path.join('results','dataset_normalizado.pkl'))
df_bankruptcy.shape
```

```python
df_bankruptcy.head(5)
```

#### 5.1. Particion Muestral

```python
dataset = df_bankruptcy
#Separamos la data, en variables independientes (x) y dependientes (y), para poder entrenar un árbol de clasificación
x = dataset.drop("flag", axis=1)

y = dataset["flag"]
```

```python
#Mediante el método "train_test_split" usaremos el 20% de la data para probar el modelo. El parámetro "random state" nos sirve para
#poder replicar la misma separación
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0, stratify=y)
```

```python
print(f"Train shape: {X_train.shape}")
print(f"Test shape: {X_test.shape}")
```

#### 5.2. Algoritmos Machine Learning

```python
#Entrenamos el árbol de clasificación
model = DecisionTreeClassifier(max_depth=3,random_state=0,criterion="gini",splitter="best")
model.fit(X_train, y_train)
```

```python
#medimos el accuracy de test
y_pred = model.predict(X_test)
predictions = y_pred
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.4f%%" % (accuracy * 100.0))
```

```python
#medimos el accuracy de train
y_pred = model.predict(X_train)
predictions = y_pred
accuracy = accuracy_score(y_train, predictions)
print("Accuracy: %.4f%%" % (accuracy * 100.0))
```

```python
viz = ConfusionMatrix(model)
viz.score(X_test, y_test)
viz.draw()
viz.finalize()
```

```python
# Curva ROC- AUC
from yellowbrick.classifier.rocauc import roc_auc
roc_auc(model, X_train, y_train, X_test=X_test, y_test=y_test, classes=['not_defaulted', 'defaulted'])
```

#### 5.2. Tunning de Algoritmos

```python
from sklearn.model_selection import GridSearchCV
```

```python
max_depths = np.arange(4, 12, 1)
scores = []
for n in max_depths:
    model.set_params(max_depth=n)
    model.fit(X_train, y_train)
    scores.append(model.score(X_test, y_test))
plt.title("Effect of max_depth")
plt.xlabel("max_depth")
plt.ylabel("score")
plt.plot(max_depths, scores)
```

```python
##Se observa una mejora en Max_depth = 5 y 6
#Entrenamos el árbol de clasificación
model = DecisionTreeClassifier(max_depth=5,random_state=0,criterion="gini",splitter="best")
model.fit(X_train, y_train)
```

```python
#medimos el accuracy de test
y_pred = model.predict(X_test)
predictions = y_pred
accuracy = accuracy_score(y_test, predictions)
print("Accuracy Test: %.4f%%" % (accuracy * 100.0))
#medimos el accuracy de train
y_pred = model.predict(X_train)
predictions = y_pred
accuracy = accuracy_score(y_train, predictions)
print("Accuracy Train: %.4f%%" % (accuracy * 100.0))
```

```python
train = dataset
target = 'flag'
predictors = [x for x in train.columns if x not in [target]]
```

```python
param_test1 = {
 'max_depth':range(1,7,1)
}
gsearch1 = GridSearchCV(estimator = DecisionTreeClassifier(random_state=0,criterion="gini",splitter="best"), 
 param_grid = param_test1, scoring='roc_auc',n_jobs=-1, cv=5)
gsearch1.fit(train[predictors],train[target])
gsearch1.cv_results_, gsearch1.best_params_, gsearch1.best_score_
```

```python
##Sin embargo se observa que el mejor valor para el max_depth=4
#Entrenamos el árbol de clasificación final
model_DT = DecisionTreeClassifier(max_depth=4,random_state=0,criterion="gini",splitter="best")
model_DT.fit(X_train, y_train)
```

```python
#medimos el accuracy de test
y_pred = model_DT.predict(X_test)
predictions = y_pred
accuracy = accuracy_score(y_test, predictions)
print("Accuracy Test: %.4f%%" % (accuracy * 100.0))
#medimos el accuracy de train
y_pred = model_DT.predict(X_train)
predictions = y_pred
accuracy = accuracy_score(y_train, predictions)
print("Accuracy Train: %.4f%%" % (accuracy * 100.0))
```
